```python
from PIL import Image, ImageGrab
import numpy as np
from numpy import asarray
import time
import pickle
import math
import keyboard
import tensorflow as tf
import cv2
```

```python
def whichKey():
    try:  # used try so that if user pressed other than the given key error will not be shown
        if keyboard.is_pressed('down'):  # if key 'q' is pressed 
            print('1')
            return 1
        elif keyboard.is_pressed('up'):
            print('2')
            return 2
        elif keyboard.is_pressed('q'):
            print('3')
            return 3
        else:
            print('0')
            return 0
    except:
        print('error!')
```

```python
m = tf.keras.models.Sequential()

m.add(tf.keras.layers.Conv2D(32,(5,5),activation='relu',input_shape=(300,300,3)))
m.add(tf.keras.layers.MaxPool2D(pool_size=(2,2)))
m.add(tf.keras.layers.Conv2D(32,(3,3),activation='relu'))
m.add(tf.keras.layers.MaxPool2D(pool_size=(2,2)))
m.add(tf.keras.layers.Flatten())
m.add(tf.keras.layers.Dense(128))
m.add(tf.keras.layers.Dense(64))
m.add(tf.keras.layers.Dense(3,activation='softmax'))

m.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
x = []
y = []
print('Game starting in 2 seconds...')
time.sleep(2)
while True:
    key = whichKey()
    if key == 3:
        break
    flatten_image = []
    image = ImageGrab.grab()
    left = 0
    top = 312
    right = 300
    bottom = 612
    image = image.crop((left, top, right, bottom))
    image.save('img.jpg')
    i = cv2.imread('img.jpg')
    x.append(i)
    y.append(key)
X = np.array(x,dtype=np.float32)
Y = np.array(y,dtype=np.float32)
no_of_images = X.shape[0]
batch_size = 150
iterations = int(math.floor(no_of_images/batch_size))
for i in range(iterations):
    start = i*batch_size
    end = (i+1)*batch_size
    X_minibatch = X[start:end,:]
    Y_minibatch = Y[start:end]
    m.fit(X_minibatch,Y_minibatch,epochs=3)
    
# file = open('modelTF.pkl','wb')

#dump info into that file:
# pickle.dump(m,file)
# file.close()
```

```python
from PIL import Image, ImageGrab
import numpy as np
from numpy import asarray
import time
import pickle
import pyautogui
import keyboard 
```

```python
def hit(key):
    pyautogui.keyDown(key)
    return

def quitKey():
    try:  # used try so that if user pressed other than the given key error will not be shown
        if keyboard.is_pressed('q'):
            print('Quitting Game...')
            return 3
    except:
        pass
```

```python
# file = open('modelTF.pkl','rb')
# clf=pickle.load(file)
# file.close()
print('Game starting in 2 seconds...')
time.sleep(2)
while True:
    if(quitKey() == 3):
        break
    else:
        image = ImageGrab.grab()
        left = 0
        top = 312
        right = 300
        bottom = 612
        image = image.crop((left, top, right, bottom))
        image.save('imggame.jpg')
        i = cv2.imread('imggame.jpg')
        x = np.array([i],dtype=np.float32)
        model_prediction = m.predict(x)[0]
        P = model_prediction/max(model_prediction)
        if P[1] ==1.0:
            print('down')
            hit("down")

        elif P[2] == 1.0:
            print('up')
            hit("up")
print('Thanks for playing the game :)')
```
